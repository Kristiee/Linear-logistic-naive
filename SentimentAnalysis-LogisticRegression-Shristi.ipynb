# 2. Feature Extraction and Feature Normalization

###### Split data set to train, validation and testing set

```python
# 60% train set
df_train = df[:train]

#20% validation set 
df_val = df[train:train+val]

#20% test set
df_test = df[train+val :]

print(len(df_train), len(df_val), len(df_test))
```

###### create vocab from train set using uniq words

```python
df_train.Phrase
```

```python
vocab = list()  

for i in df_train.Phrase.tolist():
    # extend used in place of append to get all words as list and prevent nested list
    vocab.extend(i.split(" "))  


#to get uniq words from vocab use set action
u_vocab = set(vocab)  
print(len(vocab), len(u_vocab))
```

```python
%time
nplist=np.array(vocab)
nplist
```

##### Feature extraction

```python
#RUle 1 using binary method and converting each sentence on phrase to binary

def extract_feature(f_df):
    feature = []
    label = []
    for index, row in df_train.iterrows():
        words = row['Phrase'].split(" ")
        f = []
        for x in u_vocab:
            if x in words:
                f.append(1)
            else:
                f.append(0)
        if row['label'] == "Positive":
            label.append(1)
        elif row['label'] == "Negative":
            label.append(0)

        feature.append(f)


    feature = np.array(feature)
    label = np.array(label)
    label = label.reshape(-1,1)
    return feature, label


X_train, Y_train = extract_feature(df_train)
X_val, Y_val = extract_feature(df_val)
X_test, Y_test = extract_feature(df_test)
print(X_train.shape)
print(Y_train.shape)
```

###### Add Bias

```python
X_train = np.insert(X_train, 0 , values = 1, axis=1)
X_val = np.insert(X_val, 0 , values = 1, axis=1)
X_test = np.insert(X_test, 0 , values = 1, axis=1)
```

```python
print(X_train.shape)
print(Y_train.shape)
```

# 3. Grid search parameter with val loss

### SIgmoid Function

```python
def sigmoid(z):
    return  1/(1 + np.exp(-z))
```

### Cost Entropy loss Function

```python
def cross_entropy_loss(y_pred, target):
    return (-target*np.log(y_pred) - (1-target)*np.log(1-y_pred)).mean()
```

###### Hyperparameter tuning

```python
np.random.seed(42)

train_error=[]


params ={
    'num_of_iteration': 5000,
    'learning_rate': 0.9,
    'regularization': 0.001

}

grid_param ={
    'num_of_iteration': [500, 4000],
    'learning_rate': [0.01, 0.8, 2],
    'regularization': [0.01, 0.001]
}
```

```python
import itertools

grid=list(itertools.product(grid_param['num_of_iteration'],grid_param['learning_rate'],grid_param['regularization']))
print(grid)
```

# 4. Train loss plot in the best model [best hyper-parameters and parameters]


```python
def train_model(X_train, y_train, X_Val, y_val,  param):
    np.random.seed(42)
    W = np.random.rand(X_train.shape[1]).reshape(-1,1)

    train_error = []
    val_error = []
    print("params", param)
    
    for i in range(1, param['num_of_iteration']):
        z = np.matmul(X_train, W)
        h_train = sigmoid(z)
        
        #calculation of regularization
        R=param['regularization']*sum(W**2)
        
        train_loss= cross_entropy_loss(h_train, y_train) #+ R

#         if i%10==0:
        train_error.append(train_loss)
            
            
        grad = h_train -y_train
        gradient_main = np.matmul(np.transpose(X_train), grad)/X_train.shape[0]
#         grad_bias = np.average(grad)
        gradient_regularization = 2*param['regularization']*W
        gradient_regularization[0]=0
        gradient=gradient_main+gradient_regularization

        W = W - param['learning_rate'] * gradient
    v = np.matmul(X_val, W)
    h_val=sigmoid(v)
    val_loss = cross_entropy_loss(h_val, y_val)
    val_error.append(val_loss)
    print("Train loss:", train_loss)
    print("Val loss:", val_loss)

    plt.plot(val_error)
    plt.xlabel('Iteratopns(per tens)')
    plt.ylabel('Validation Cost')
    plt.title('Learning Rate')
    plt.show()
    return W, train_error, val_error
```

##### Model fine tuning

```python
for g in grid:
    p={
    'num_of_iteration':g[0],
    'learning_rate':g[1],
    'regularization':g[2]
    }
    Weights,train_error,val_error=train_model(X_train,Y_train,X_val,Y_val,p)
#     print(train_error)
    plt.plot(np.arange(len(train_error)),train_error)
    plt.xlabel("No. of iterations")
    plt.ylabel("Training Error")   
    plt.show()
```

```python
Weights,train_error,val_error=train_model(X_train,Y_train,X_val,Y_val,params)
```
